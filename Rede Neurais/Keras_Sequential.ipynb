```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical #ao invez de usar keras.utils np_utils
```

```python
base = datasets.load_iris()
previsores = base.data
classe = base.target
classe
```

```python
#É necessário transformar a classe em formato dummy pois teremos 3 neuronios na camada de saida
classe_dummy = to_categorical(classe)
classe_dummy
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(previsores, classe_dummy, test_size = 0.3, random_state = 0)
```

```python
#Criaçao da estrutura da rede neural com a classe sequential (sequencia de camadas)
#Siginifca que as camadas serao incluidas uma apos a outra
modelo = Sequential()
#Primeira camada oculta, 5 neuronios, 4 neuronios de entrada
#'Dense' signifca que todos os neuronios sao conectados aos da proxima camada
#'Units' é a quantidade de neuronio a serem adicionados
#'Input_dim' é o parametro que informa a quantidade de neuronios na camada de entrada
modelo.add(Dense(units = 5, input_dim = 4))
#Segunda camada
modelo.add(Dense(units = 4))
#Terceira camada
#Funçao softmax pois há um problema de classificaçao com mais de duas classe
#É gerado uma probabilidade em cada neuronio
modelo.add(Dense(units = 3, activation = 'softmax'))
```

```python
modelo.summary()
#Camadas e tipos | quantidade de neuronios | quantidade de hiperparametros
```

```python
#Configurando os parameteos da rede neural (adam = algoritimo para atualizar os pesos (mais comum), loss = calculo do erro
modelo.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
#treinamento, dividindo a base de treinamento em uma porcaopara validacao (validation_data)
#Ele ajusta os pesos de acordo com os dados de teste validando os dados com os de teste
modelo.fit(x_treino, y_treino, epochs = 1000, validation_data = (x_teste, y_teste))
#accuracy = valor com dados de treino, val_accuracy = é o mais importante pois é utilizado os dados de teste, ou seja, ele nao usou no treino
#loss = é o quanto está errando
#O mais importante é o val_accuracy, pois melhor demonstra a realidade, neste caso o modelo teve uma taxa de 97% de acerto
```

```python
#Realizar previsoes e alterar o valor de saida para true or false para poder utilizar na matriz confusao
previsao = modelo.predict(x_teste)
previsao = (previsao > 0.5)
previsao
```

```python
#Buscando o maior valor
y_teste_matrix = [np.argmax (t) for t in y_teste]
y_previsao_matrix = [np.argmax (t) for t in previsao]
```

```python
#Matriz de confusao
confusao = confusion_matrix(y_teste_matrix, y_previsao_matrix)
confusao
```
