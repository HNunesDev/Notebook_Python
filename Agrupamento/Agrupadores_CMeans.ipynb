```python
from sklearn import datasets
import numpy as np
from sklearn.metrics import confusion_matrix
import skfuzzy
```

```python
iris = datasets.load_iris()
iris
```

```python
agrupamento = skfuzzy.cmeans(data = iris.data.T, c = 3, m = 2, error = 0.005, maxiter = 1000, init = None)
```

```python
previsoes_porcentagem = agrupamento[1]
```

```python
for i in range(150):
    print(previsoes_porcentagem[0][i],previsoes_porcentagem[1][i],previsoes_porcentagem[2][i])
```

```python
previsao = previsoes_porcentagem.argmax(axis = 0)
resultados = confusion_matrix(iris.target, previsao)
resultados
```
